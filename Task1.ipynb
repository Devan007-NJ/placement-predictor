#**DELIVERABLE 2: Building the Machine Learning Model**

##Spliting data to training and testing

```python
x=df.drop(columns=['StudentID','PlacementStatus'],axis=1)
y=df['PlacementStatus']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
x_train.shape,x_test.shape
```

##Creating and Training the model

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=200, # number of trees
    max_depth=4, # max depth of the decision tree classifier
    learning_rate=0.1,
    subsample=0.9, # Fraction of training rows used to build each tree. reduces overfitting
    colsample_bytree=0.9,
    eval_metric='logloss'
)

model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

#**DELIVERABLE 3: TO CHECK THE ACCURACY OF THE MODEL**

##Checking accuracy using metrics

```python
from sklearn.metrics import  mean_squared_error,r2_score
model_mse=mean_squared_error(y_test,y_pred)
model_r2=r2_score(y_test,y_pred)
print("MSE:",model_mse)
print("R2:",model_r2)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```

```python
print("Accuracy in percentage: 79.75%")
```

##Comparison of model vs actual values by visualization

```python
mapping = {0: "No", 1: "Yes"}
y_test_mapped = pd.Series(y_test).map(mapping)
y_pred_mapped = pd.Series(y_pred).map(mapping)

# Combine into a single dataframe
frame = pd.DataFrame({
    "Value": list(y_test_mapped) + list(y_pred_mapped),
    "Dataset": ["y_test"] * len(y_test) + ["y_pred"] * len(y_pred)
})
sns.countplot(data=frame, x="Value", hue="Dataset")
plt.title("Comparison of Actual vs Predicted Values")
plt.ylabel("Count")
plt.xlabel("Class (No/Yes)")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True,fmt='0.2f',cmap='Reds')
plt.xlabel("Predicted values")
plt.ylabel("Actual values")
```

```python
import shap
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(x_test)
shap.summary_plot(shap_values, x_test,plot_type='bar')
```
